```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
plt.rc('font',family='Times New Roman') 
```

```python
def listdir(path, list_name):
    for file in os.listdir(path):
        file_path = os.path.join(path, file)
        if os.path.isdir(file_path):
            listdir(file_path, list_name)
        else:
            list_name.append(file_path)
    return list_name
```

```python
all_file = []
all_file = listdir(r".//finaladdboxplot", all_file)
all_filename = []
for i in all_file:
    temp_filename = i.split("\\")
    temp_filename = temp_filename[-1].split(".")[0]
    all_filename.append(temp_filename)
print(all_filename)
labels = ['Fuzzy control', 'PID-Modified RNN', 
          'PID-Original RNN', 'PID-PSO', 
          'PID-SOA', 'Fuzzy control', 
          'PID-Modified RNN', 'PID-Original RNN', 
          'PID-PSO', 'SOA']
```

```python
datalist = []
for i in range(len(all_filename)):
    readfile = pd.read_csv(r".//finaladdboxplot//" + all_filename[i]+".csv")
    readfile = readfile["edu"].tolist()

    datalist.append(readfile)
    
```
