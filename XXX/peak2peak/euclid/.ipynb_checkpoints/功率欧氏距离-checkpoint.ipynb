```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
def listdir(path, list_name):
    for file in os.listdir(path):
        file_path = os.path.join(path, file)
        if os.path.isdir(file_path):
            listdir(file_path, list_name)
        else:
            list_name.append(file_path)
    return list_name
```

```python
all_file = []
all_file = listdir(r".//edu2", all_file)
all_filename = []
for i in all_file:
    temp_filename = i.split("\\")
    temp_filename = temp_filename[-1].split(".")[0]
    all_filename.append(temp_filename)
print(all_filename)
labels = ['nonvmd_nn_new', 'nonvmd_nn_old', 'nonvmd_pso', 
          'nonvmd_soa', 'vmd_nn_new', 'vmd_nn_old', 
          'vmd_pso', 'vmd_soa']
```

```python
datalist = []
for i in range(len(all_filename)):
    readfile = pd.read_csv(r".//edu//" + all_filename[i]+".csv")
    readfile = readfile["edu"].tolist()

    datalist.append(readfile)
    
```

```python
plt.figure(dpi=600)
plt.figure(figsize=(15, 12))
plt.rcParams['savefig.dpi'] = 600 #
plt.rcParams['figure.dpi'] = 600 #
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False

# N1np
plt.subplot(4, 1, 1)
# plt.ylim(-50,30) 
# plt.legend("N1",fontsize='x-large', loc=1)
plt.title("N1", fontdict={'weight':'normal','size': 30})
# plt.boxplot(datalist, labels = labels,
#             notch = True,sym = '*', patch_artist = True, boxprops = {'color':'orangered','facecolor':'green', 'alpha':0.7})
plt.boxplot(datalist, labels = labels, 
            notch = True,sym = '*')
# plt.tick_params(
#     axis='x',          # changes apply to the x-axis
#     which='both',      # both major and minor ticks are affected
#     bottom=False,      # ticks along the bottom edge are off
#     top=False,         # ticks along the top edge are off
#     labelbottom=False) # labels along the bottom edge are off
plt.xticks(fontsize=23, rotation = 45)
plt.yticks(fontsize=16)
axes = plt.gca()
left, right = axes.get_xlim()

# axes.hlines(y=10, xmin=left, xmax=right, linestyles='dashed')
# axes.hlines(y=4, xmin=left, xmax=right, linestyles='dashed')
# axes.hlines(y=19, xmin=left, xmax=right, linestyles='dashed')
```
