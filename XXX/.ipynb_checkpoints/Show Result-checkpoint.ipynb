```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import math
from numpy import array
import os
```

```python
def dataread(path):
    df = pd.read_csv(path)
    timek = df['time'].values.copy()
    valuek = df['value'].values.copy()
    return timek,valuek
```

```python
name = {0:'soa',1:'nn_old',2:'pso',3:'nn_new'}
vmd=  {}
nonvmd ={}


sample,original = dataread("./datasources/validation/original.csv")
plt.figure(figsize=(8,8))
count = 1
for i in range(4):
    
    sample,vmd[i] = dataread( "./Vmd/Data/{}.csv".format(name[i]))
    plt.subplot(5,2,count)
    plt.title('VMD:{} & Original'.format(name[i]) )
    plt.plot(sample,original,label='orig')
    plt.plot(sample,vmd[i],label='{}'.format(name[i]))
    plt.legend(loc=2)
    
    sample,nonvmd[i] = dataread( "./Nonvmd/Data/{}.csv".format(name[i]))
    plt.subplot(5,2,count+1)
    plt.title('nonVMD:{} & Original'.format(name[i]) )
    plt.plot(sample,original,label='orig')
    plt.plot(sample,nonvmd[i],label='{}'.format(name[i]))
    plt.legend(loc=2)

    count = count + 2
    
sample,nonfuzzy = dataread( "./Nonvmd/Data/fuzzy.csv".format(name[i]))
plt.subplot(5,2,10)
plt.title('nonVMD:fuzzy & Original')
plt.plot(sample,original,label='orig')
plt.plot(sample,nonfuzzy,label='fuzzy')
plt.legend(loc=2)

sample,nonfuzzy = dataread( "./Vmd/Data/fuzzy.csv".format(name[i]))
plt.subplot(5,2,9)
plt.title('VMD:fuzzy & Original')
plt.plot(sample,original,label='orig')
plt.plot(sample,nonfuzzy,label='fuzzy')
plt.legend(loc=2)

plt.tight_layout()
```
